**2. Scan all numeric variables for outliers. If there are outliers, use any of the suitable
techniques to deal with them**

Outliers can be defined as values that are unusually very large or small compared to the other values in the dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
import io
data=files.upload()
```

```python
df=pd.read_csv("stdheight.csv")
df
```

```python
df.describe()
```

```python
x=df['Height']
plt.hist(x)
plt.show()
```

Q1 and Q3 refer to the first and third quartiles, respectively, of a dataset. The quartiles divide a dataset into four equal parts, with Q1 representing the 25% and Q3 representing the 75% of the data in the 'Height' column

```python
Q1=df.Height.quantile(0.25)
Q3=df.Height.quantile(0.75)
Q1,Q3
```

Interquartile range

```python
IQR=Q3-Q1
IQR
```

```python
lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR
lower_limit , upper_limit
```

Outliers

```python
df[(df.Height<lower_limit)|(df.Height>upper_limit)]
```

Remove Outliers

```python
df_no_outliers=df[(df.Height>lower_limit)&(df.Height<upper_limit)]
df_no_outliers
```

```python
x=df_no_outliers['Height']
plt.hist(x)
plt.show()
```

```python
import pandas as pd
import numpy as np
```

```python
data={'Name':['A','B','C','D','E'],
       'Age':[18,19,17,19,20],
      'Gender':['Male','Female','Female','Male','Female'],
      'Science':[80,87,65,98,67],
      'Attendance':[90,50,95,60,100]
      }
df=data
df
```

```python
df=pd.DataFrame(data)
df
```

Log Tranformation

```python
df['Attendance_log']=np.log(df['Attendance'])
#Chek the distribution before and after transformation
import matplotlib.pyplot as plt
fig,axs=plt.subplots(1,2,figsize=(10,5))
axs[0].hist(df['Attendance'],bins=30)
axs[0].set_title('Attendence distribution (Before Transformation)')
axs[1].hist(df['Attendance_log'],bins=30)
axs[1].set_title('Attendence distribution (After Transformation)')
plt.show()
```
